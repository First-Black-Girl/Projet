```python
import pandas as pd

# Chemin du fichier Excel
fichier_excel = "Data_project.xlsx"

# Lire toutes les feuilles dans un dictionnaire
donnees_excel = pd.read_excel(fichier_excel, sheet_name=None)

# Parcourir les feuilles et afficher les données
for nom_feuille, dataframe in donnees_excel.items():
    print(f"Feuille : {nom_feuille}")
    print("\n")
```

```python
for nom_feuille, dataframe in donnees_excel.items():
    fichier_csv = f"{nom_feuille}.csv"
    dataframe.to_csv(fichier_csv, index=False)
    print(f"Feuille {nom_feuille} enregistrée dans {fichier_csv}")
```
